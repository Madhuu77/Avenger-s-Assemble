```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame
df = pd.DataFrame({
    "date": pd.date_range(start="1990-09-07", periods=25, freq="YE"),
    "temp": np.random.randint(18, 30, size=25)
})

print(df)

# Plot the DataFrame
plt.figure(figsize=(10, 5))
ax = df.plot(kind='bar', x='date', y='temp', color='g', linewidth=1,figsize=(10,6))
plt.xticks(rotation=25)
ax.set_ylabel('Temperature')
ax.set_xlabel('Date')
plt.title('Temperature over Time')
plt.grid(True)
plt.show()
```

#Titanic Dataset


```python
import pandas as pd
dataframe=pd.read_csv("C:/Users/91861/Desktop/tested.csv")
print(dataframe)
```

```python
import matplotlib.pyplot as plt
age_survived=dataframe[dataframe['Survived']==1]['Age']
age_not_survived=dataframe[dataframe['Survived']==0]['Age']
plt.hist(age_survived,color='r',alpha=0.5,label='Surived ')
plt.hist(age_not_survived,color='g',alpha=0.5,label='Not Survived')
plt.title("Suvival of Passenger's of Titanic Ship on the basis of Age")
plt.xlabel('Age')
plt.ylabel('Freequency')
plt.legend()
plt.show()
```

```python
p_class=dataframe["Pclass"].value_counts()
plt.bar(p_class.index,p_class.values,width=0.5,color=['r','g','b'])
plt.xticks([1,2,3],['1st Class','2nd Class','3rd Class'],rotation=45)
plt.xlabel('Classes')
plt.ylabel('Survived')
plt.title('Survival rate of People on the basis of their classes')
plt.show()
```

```python
c=dataframe["Pclass"].value_counts()
co=['g','r','k']
plt.bar(c.index,c.values,color=co,width=0.5)
plt.xticks([1,2,3],["1st class","2nd class","3rd class"])
plt.xlabel("Classes");plt.ylabel("No of Passengers");plt.title("No of Passengers Travelled in Specific Classes")
plt.show()
```

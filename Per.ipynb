```python
import pandas as pd

DF=pd.read_csv("C:/Users/91861/Documents/bd-dec22-births-deaths-by-region.csv")
print(DF)
```

```python

# Filter data for 2005 and specific regions
data_count = DF[(DF['Period'] == 2022) & (DF['Region'].isin(['Auckland region', 'New Zealand']))]['Count']
data_region = DF[(DF['Period'] == 2022) & (DF['Region'].isin(['Auckland region', 'New Zealand']))]['Region']

import matplotlib.pyplot as plt
plt.bar(data_region,data_count)
plt.title('')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
df=DF.groupby('Birth_Death')['Count'].sum()
plt.pie(df,labels=df.index,autopct='%2.2f%%',explode=[0,0.1],shadow=True)
plt.legend(['Births','Deaths'])
plt.title('Comparision on No. of deaths and Births in Specific regions in 2022') 
plt.show()
```

```python
aggregated_data = DF.groupby('Birth_Death')['Count'].sum()

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(aggregated_data, labels=aggregated_data.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Births and Deaths')
plt.show()
```

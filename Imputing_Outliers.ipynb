```python
#Imputing mean to the outliers


import pandas as pd
import numpy as np
df=pd.read_csv("C:/Users/91861/Desktop/tested.csv")
tic=df['Fare']
df
q1=tic.quantile(0.25)
q2=tic.quantile(0.75)
iqr=q2-q1
low=q1-1.5*iqr
high=q2+1.5*iqr
outlier=(tic<=low)|(tic>=high)

print('No of Outlier:',outlier.sum())
mean=tic[(tic>=low)&(tic<=high)].mean()
print(mean)

#imputing mean
df.loc[outlier,'Fare']=mean

out=(df['Fare']<=low)&(df['Fare']>=high).sum()
print('Outliers after imputing:',out)

print(df.describe())

import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(df['Fare'])
plt.show()
```

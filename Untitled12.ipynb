```python
#Entering data
Employee={
    'E_name':['Roshan','Amar','Ashwini','Lohith','Mohan','pramod'],
    'Type':['Regular','Adhoc','Regular','Adhoc','Contarct','Regular'],
    'Department':['Cs','Cs','Ec','EC','Cs','Ec'],
    'Expierience':[10,20,5,14,9,8],
    'Salary':[50000,15000,30000,10000,4000,10000]
}
```

```python
#Creating Dataframe
df2=pd.DataFrame(Employee)
print(df2)
```

```python
#Average salary of Employees in each department
pivot=pd.pivot_table(df2,index='Department',values='Salary',aggfunc='mean')
print(pivot)
```

```python
#Average sum and count of each type of employee
pivot2=pd.pivot_table(df2,index='Type',values='Salary',aggfunc=['mean','sum','count'])
print(pivot2)
```

```python
#Plotting average salary of each Department
import matplotlib as plt

s=pivot.plot(kind='bar',figsize=(10,6))
```

```python
#Plotting for Average salary of Each Type of Employee
pivot2.plot(kind='bar',figsize=(10,6))
```

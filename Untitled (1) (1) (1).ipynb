pivot in pandas

# Assistant
In pandas, the `pivot` function is used to reshape a DataFrame by pivoting the data from rows into columns. This function is particularly useful when you want to reorganize your data to make it more readable or to perform further analysis.

The `pivot` function takes three main arguments:
- `index`: The column to use as the new index.
- `columns`: The column to use as the new columns.
- `values`: The column to use as the new values.

Here's an example of how to use the `pivot` function in pandas:


```python
import pandas as pd

# Create a sample DataFrame
data = {
    'Date': ['2021-01-01', '2021-01-01', '2021-01-02', '2021-01-02'],
    'City': ['New York', 'Los Angeles', 'New York', 'Los Angeles'],
    'Temperature': [32, 75, 30, 68],
    'Humidity': [60, 40, 75, 55]
}

df = pd.DataFrame(data)

# Use pivot to reshape the DataFrame
pivot_df = df.pivot(index='Date', columns='City', values='Temperature')

print(pivot_df)
```


In this example, we have a DataFrame with columns for Date, City, Temperature, and Humidity. We use the `pivot` function to pivot the data so that the Date becomes the index, the City becomes the columns, and the Temperature becomes the values. The resulting DataFrame will have Dates as the index, Cities as the columns, and Temperature values at the intersections.

The `pivot` function is a powerful tool for reshaping data in pandas, and it can be used in various scenarios to make the data more accessible for analysis.

```python
import pandas as pd

# Step 1: Create the initial DataFrame
data = {
    'ID': [100, 110, 120, 130],
    'State': ['Delhi', 'Mumbai', 'Chennai', 'Surat'],
    'Cases': [3000, 4000, 5000, 4500]
}
df = pd.DataFrame(data, index=[100, 110, 120, 130])
print("Initial DataFrame:\n", df)

# Step 2: Add the "Recovery" column using the Series method
recovery_data = pd.Series([2500, 3500, 4500, 4000], index=[100, 110, 120, 130])
df['Recovery'] = recovery_data
print("\nDataFrame with Recovery column:\n", df)

# Step 3: Add the "Deaths" column using the `assign()` method
df = df.assign(Deaths=[100, 200, 300, 250])
print("\nDataFrame with Deaths column:\n", df)

# Step 4: Add a new row to store details of another state using `loc`
df.loc[140] = [140, 'Bangalore', 6000, 5500, 400]
print("\nDataFrame with new row added:\n", df)

# Step 5: Add a "Percentage" column using the `insert()` method
df.insert(3, 'Percentage', [83.33, 87.5, 90, 88.89, 91.67])
print("\nDataFrame with Percentage column inserted:\n", df)

# Step 6: Delete the "Percentage" column using `del`
del df['Percentage']
print("\nDataFrame after deleting Percentage column:\n", df)

# Step 7: Delete the "Deaths" column using `pop()`
df.pop('Deaths')
print("\nDataFrame after deleting Deaths column:\n", df)

# Step 8: Insert a new row of values using `iloc` at the 1st position
df = pd.concat([df.iloc[:1], pd.DataFrame([[101, 'Hyderabad', 3500, 3000]], columns=df.columns, index=[101]), df.iloc[1:]]).reset_index(drop=True)
print("\nDataFrame after inserting new row at the 1st position:\n", df)

# Step 9: Temporarily delete "Cases" and "State" columns
temp_df = df.drop(columns=['Cases', 'State'])
print("\nDataFrame after temporarily deleting 'Cases' and 'State' columns:\n", temp_df)

# Restoring the original DataFrame
print("\nRestoring the original DataFrame:\n", df)
```

```python
#Training model and predicting Salary of employee based on experience using Linear_Regression.
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error,accuracy_score

df=pd.read_csv("C:/Users/91861/Downloads/Salary.csv")
df.info()

x=df['YearsExperience'].values.reshape(-1,1)
y=df['Salary']


#Using Linear Regression to get output for given input and also train data.
LR=LinearRegression()
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)

#Fitting the training set into Linear Regression model for Training model on input and output.
LR.fit(x_train,y_train)

#Predicting Values using Test set which is used to test the model accuracy
y_pred=LR.predict(x_test)

#Accuracy of Linear Regression model
#Linear_Regression_Accuracy=accuracy_score(y_test,y_pred)
#print(Linear_Regression_accuracy)
mse=mean_squared_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
print("\nMean Sqaured error :",mse)
print("R^2 Score :",r2)

#Plotting Testing and predicted Data using scatter and line plot
import matplotlib.pyplot as m
m.figure(figsize=(10, 6))
m.scatter(x_test, y_test, color='green', label='Actual Data')
m.plot(x_test, y_pred, color='y', linewidth=2, label='Regression Line')
m.title('Linear Regression: Salary vs. Years of Experience')
m.xlabel('Years of Experience');m.ylabel('Salary')
m.legend();m.show()


#Prediction of Salary of Expierience 15 years
exp=[[15]]
pred_salary=LR.predict(exp)

print(f'Salry of Employee {exp[0][0]} Year Expirience:{pred_salary[0]}')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score,mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

#Loading Data from csv file
df=pd.read_csv('C:/Users/91861/Downloads/Student_Marks.csv')
print9

#Assigning Columns to variables and reshaping time study column to 2D
x=df['time_study'].values.reshape(-1,1)
y=df['Marks']


#splitting data into training and testing set int the ratio of 7:3.
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=100)

#Using Linear Regression model to train the data and make prediction to training data.
Lr=LinearRegression()
Lr.fit(x_train,y_train)
y_pred=Lr.predict(x_test)

#Using metrics to Evaluate the efficiency of Prediction
mse=mean_squared_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
print('Mean Squared error:',mse)
print('R^2 Score:',r2)

#Plotting the Testing and Predicted data using scatter and line plot
plt.scatter(x_test,y_test,c='g',label='Actual data')
plt.plot(x_test,y_pred,c='r',label='Regressor Line')
plt.xlabel('Study Hours')
plt.ylabel('Marks Obtained')
plt.legend()
plt.title('Tested data vs Predicted data')
plt.show()

#Predicting Marks on the study  hours of 15h.
hours=[[15]]
marks_pred=Lr.predict(hours)
print(f'\n->Prediction\nThe marks of Student {hours[0][0]} of study :{marks_pred[0]}')
```

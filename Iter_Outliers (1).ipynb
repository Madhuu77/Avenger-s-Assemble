```python
from sklearn.datasets import load_diabetes
import pandas as pd
dia=load_diabetes()
col=dia.feature_names
df=pd.DataFrame(data=dia.data,columns=col)
df
```

```python
#Plotting Box plot for Each feature using iterative statement
import seaborn as sns
import matplotlib.pyplot as plt

fig, axs = plt.subplots(ncols=5, nrows=2, figsize=(20, 10))
index = 0
axs = axs.flatten()
for k,v in df.items():
    sns.boxplot(y=k, data=df, ax=axs[index])
    index += 1
plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=5.0)
```

```python
#Findinng out Outliers using For Loop.
import numpy as np
for k,v in df.items():
    q1=v.quantile(.25)
    q2=v.quantile(.75)
    iqr=q2-q1
    v_col=v[(v<=q1-1.5*iqr)|(v>=q2+1.5*iqr)]
    perc=np.shape(v_col)[0]*100/np.shape(df)[0]
    print("Column %s Outliers perc:%.2f%%"%(k,perc))
```

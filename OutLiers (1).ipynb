```python
from sklearn.datasets import load_diabetes
import matplotlib.pyplot as m
import pandas as p
import seaborn as s
dp=load_diabetes()
col_n =dp.feature_names
df= p.DataFrame(dp.data);df.columns = col_n
m.figure(figsize=(10,6))
s.boxplot(df['bmi'])
m.xlabel('bmi')
m.ylabel('Values')
m.title('Distribution of BMI')
m.show()

#IQr 
q1=df['bmi'].quantile(0.25)
q2=df['bmi'].quantile(0.75)
iqr=q2-q1

#Floor and capping
flo=q1-1.5*iqr
cap=q2+1.5*iqr
out=df[(df['bmi']<=flo)|(df['bmi']>=cap)]
print("Outliers:-",out)
```

```python
from sklearn.datasets import load_diabetes
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

dd=load_diabetes()
col=dd.feature_names
df1=pd.DataFrame(dd.data)
df1.columns=col

plt.figure(figsize=(10,6))
sns.boxplot(data=df1)
plt.xlabel('Columns')
plt.ylabel('values')
plt.title('Visualizing outliers using Box plot')


```

```python
#Major Reason For Finding or Determination of outlier....
#As you can see here that the Outliers are the Vlaues which are Extremely larger or smaller than the Average values in the Dataset So it may lead to or cause Effeciency or Accuracy of the Model is trained with data can affect prediction so we need to find out the Outliers and etxract from the datasets.
```

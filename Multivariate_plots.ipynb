```python
import pandas as p
import matplotlib.pyplot as m
import seaborn as s
# data as 32 Elements
data=p.read_csv("C:/Users/91861/Desktop/auto-mpg.csv")
# HISTOGRAM
mpg=data['mpg']
m.hist(mpg,bins='auto',color='k',edgecolor='c')
m.xlabel('Miles per gallon (mpg)');m.ylabel('Frequency')
m.title('Frequency Distribution of mpg')
m.show()
# SCATTER
wt=data['weight']
iv=range(len(data))
m.scatter(iv,mpg,color='k',label='mpg')
m.scatter(iv,wt,color='g',label='weigth')
m.title("Relationship b/w Weigth and MPG")
m.legend()
m.show()

# BOX PLOT
s.boxplot(mpg,color='c')
m.xlabel("MPG");m.ylabel("Values")
m.title("BOX plot of MPG Vlues")
```

```python
#Univariate outliers


from sklearn.datasets import load_diabetes
import matplotlib.pyplot as m
import seaborn as s
dp=load_diabetes()
col_n =dp.feature_names
df= p.DataFrame(dp.data);df.columns = col_n
#Visualizing of Outliers
s.boxplot(df['bmi'])
m.ylabel('Values');m.xlabel('bmi');m.title('Distrubution of bmi')
m.show()
#IQR
q1=df['bmi'].quantile(0.25)
q3=df['bmi'].quantile(0.75)
iqr=q3-q1
#Floor and Capping
flo=q1-1.5*iqr
cap=q3+1.5*iqr
out=df[(df.bmi<=flo)|(df.bmi>=cap)]
print("Outliers:\n",out)
```

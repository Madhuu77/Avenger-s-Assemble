```python
#Import All the Required Packages for Performing Exploration,Preprocessing and Splitting Data
import pandas as pd
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Loading dataset using pandas Library of Virat Kohli Performance
df=pd.read_csv('C:/Users/91861/Downloads/Virat Kohli DataSet.csv')
df
```

```python
#transforming Data using LabelEncoder
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Dismissal']=le.fit_transform(df['Dismissal'])
df['Opposition']=le.fit_transform(df['Opposition'])
df['Ground']=le.fit_transform(df['Ground'])
df['Start Date']=le.fit_transform(df['Start Date'])
```

```python
#Finding Out the null values 
df.isna().sum()
```

```python
#Exploration Data Through Visualization
#Hist plot
plt.figure(figsize=(10,6))
sns.histplot(df['Runs'],bins=10,kde=True)
plt.title('Distribution of Runs Scored')
plt.show()
```

```python
#Bar plot
plt.figure(figsize=(12,8))
ax=sns.barplot(data=df,x='Opposition',y='Runs',palette=['b','y','g','darkgreen','brown','k','lime','orange','maroon','darkgreen','r'])
countries=['Sri Lanka','Australia','Bangladesh','South Africa','Zimbambwe','New Zealand','Ireland','Netherlands','West Indies','Pakisthan','England']
plt.xticks(rotation=15)
ax.set_xticklabels(countries)
plt.show()
```

```python
#Heatmap of kohli statistics
plt.figure(figsize=(10,6))
sns.heatmap(data=df.corr(),annot=True,cmap='crest')
plt.title('Heatmap on Kohli stats')
plt.show()
```
